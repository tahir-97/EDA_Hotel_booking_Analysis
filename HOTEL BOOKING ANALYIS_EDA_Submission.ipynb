````python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_bar_chart_from_column(df, column_label, title):
    plt.figure(figsize=(12, 6))
    sns.set(style="whitegrid")  # Set the style of the plot

    # Use Seaborn's countplot to create the bar chart
    sns.countplot(data=df, x=column_label, palette="Set2")

    plt.title(title)
    plt.xlabel(column_label)
    plt.ylabel("Count")

    plt.xticks(rotation=15)  # Rotate x-axis labels for better readability

    plt.show()
````

````python
plot_bar_chart_from_column(hotel_booking_df, "hotel", "Hotel booking by name")
````

````python
def plot_pie_chart_from_column(df, column_label, title, explode=None):
    plt.figure(figsize=(10, 6))

    # Get the counts for each category
    counts = df[column_label].value_counts()

    # If explode is specified, create a tuple for it, else create a tuple of zeros
    if explode:
        explode_tuple = tuple(explode)
    else:
        explode_tuple = tuple([0] * len(counts))

    # Create the pie chart
    plt.pie(counts, labels=counts.index, autopct='%1.2f%%', startangle=90, shadow=True, explode=explode_tuple)

    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.legend()

    plt.show()
````

````python
exp1 = [0.05,0.05]
plot_pie_chart_from_column(hotel_booking_df, 'hotel', 'Booking percentage of Hotel by Name', exp1)  # Specify explode as a list if needed
````
